```python
# Load libraries

import pandas as pd
import numpy as np
```

```python
# Import dataset

df = pd.read_csv('Telco_customer_churn.csv')

# View dataset

df.head()
```

```python
# Drop columns not required for making predictions 

df.drop(['Churn Label', 'Churn Score', 'CLTV', 'Churn Reason'],
        axis = 1, inplace = True)

df.head()
```

```python
# Dropping non-unique columns

df.drop(['CustomerID', 'Count', 'Country', 'State', 'Lat Long'],
        axis = 1, inplace = True)

df.head()
```

```python
# Formatting data for xgboost

df['City'].replace(' ', '_', regex = True, inplace = True)

df.head()
```

```python
df.columns = df.columns.str.replace(' ', '_')

df.head()
```

```python
# Checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col] == ' ')
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Checking the number of missing values in Total Charges

len(df.loc[df['Total_Charges'] == ' '])
```

```python
# Analyzing the rows that contain missing values

df.loc[df['Total_Charges'] == ' ']
```

```python
# Changing missing Total Charges to zero as their tenures are zero

df.loc[(df['Total_Charges'] == ' '), 'Total_Charges'] = 0
```

```python
df.loc[df['Tenure_Months'] == 0]
```

```python
# Changing Total Charges data type from object to float

df['Total_Charges'] = pd.to_numeric(df['Total_Charges'])

df.dtypes
```

```python
df.replace(' ', '_', regex = True, inplace = True)

df.head()
```

```python
# Creating dependent variables

X = df.drop('Churn_Value', axis = 1).copy()

X.head()
```

```python
# Seperating independent variable

y = df['Churn_Value'].copy()

y.head()
```

```python
X.dtypes
```

```python
# Test one hot encoding on categorical variable

pd.get_dummies(X, columns = ['Payment_Method']).head()
```

```python
# One hot encoding categorical columns

X_encoded = pd.get_dummies(X, columns = ['City',
                                         'Gender',
                                         'Senior_Citizen',
                                         'Partner',
                                         'Dependents',
                                         'Phone_Service',
                                         'Multiple_Lines',
                                         'Internet_Service',
                                         'Online_Security',
                                         'Online_Backup',
                                         'Device_Protection',
                                         'Tech_Support',
                                         'Streaming_TV',
                                         'Streaming_Movies',
                                         'Contract',
                                         'Paperless_Billing',
                                         'Payment_Method'])

X_encoded.head()
```

```python
y.unique()
```

```python
# Import libraries to build a model

from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
# Gaussian Naive Bayes

gnb = GaussianNB()

cv = cross_val_score(gnb, X_encoded, y, cv = 5)

print(cv)
print(cv.mean())
```

```python
# Logistic Regression

lr = LogisticRegression(max_iter = 2000)

cv = cross_val_score(lr, X_encoded, y, cv = 5)

print(cv)
print(cv.mean())
```

```python
# Decision Trees

dt = tree.DecisionTreeClassifier(random_state = 42)

cv = cross_val_score(dt, X_encoded, y, cv = 5)

print(cv)
print(cv.mean())
```

```python
# KNN Classifier

knn = KNeighborsClassifier()

cv = cross_val_score(knn, X_encoded, y, cv = 5)

print(cv)
print(cv.mean())
```

```python
# Random Forest Classifier

rf = RandomForestClassifier(random_state = 42)

cv = cross_val_score(rf, X_encoded, y,cv = 5)

print(cv)
print(cv.mean())
```

```python
# Support Vector Machines

svc = SVC(probability = True)

cv = cross_val_score(svc, X_encoded, y,cv = 5)

print(cv)
print(cv.mean())
```

```python
# Xtreme Gradient Boosting

xgb = XGBClassifier(random_state = 42)

cv = cross_val_score(xgb, X_encoded, y, cv=5)

print(cv)
print(cv.mean())
```

```python
# Function to report performance of tuned models

def clf_performance(classifier, model_name):
    print(model_name)
    print('Best Score: ' + str(classifier.best_score_))
    print('Best Parameters: ' + str(classifier.best_params_))
```

```python
# Logisitic Regression with Model Tuning

lr = LogisticRegression()

param_grid = {'max_iter' : [2000],
              'penalty' : ['l1', 'l2'],
              'C' : np.logspace(-4, 4, 20),
              'solver' : ['liblinear']}

clf_lr = GridSearchCV(lr, param_grid = param_grid, cv = 5, verbose = True, n_jobs = -1)

best_clf_lr = clf_lr.fit(X_encoded, y)

clf_performance(best_clf_lr, 'Logistic Regression')
```

```python
# KNN Classifier with Model Tuning

knn = KNeighborsClassifier()

param_grid = {'n_neighbors' : [3,5,7,9],
              'weights' : ['uniform', 'distance'],
              'algorithm' : ['auto', 'ball_tree','kd_tree'],
              'p' : [1,2]}

clf_knn = GridSearchCV(knn, param_grid = param_grid, cv = 5, verbose = True, n_jobs = -1)

best_clf_knn = clf_knn.fit(X_encoded, y)

clf_performance(best_clf_knn, 'KNN')
```
